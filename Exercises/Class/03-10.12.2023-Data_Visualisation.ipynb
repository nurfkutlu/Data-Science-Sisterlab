```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors #https://matplotlib.org/stable/gallery/color/named_colors.html
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.DataFrame({
    'Group':['A','B','C','D','E'],
    'Value':[1,5,4,3,9]


})
```

```python
# Histogram

plt.barh(y=df.Group,width=df.Value)
```

```python
#Sort
df=df.sort_values(by=['Value'])
plt.barh(y=df.Group,
         width=df.Value,
         color=['red','cyan'],
         edgecolor='black',
         linewidth=3
         )
```

```python
#Sort
#color
#edgeColor
#Linewidth
df=df.sort_values(by=['Value'])
plt.barh(y=df.Group,
         width=df.Value,
         color=mcolors.TABLEAU_COLORS, #https://matplotlib.org/stable/gallery/color/named_colors.html
         edgecolor='black',
         linewidth=3,
         )
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
sns.set(style="darkgrid")
sns.barplot(
    x='total_bill',
    y='day',
    data=tips,
    color="blue"
)
plt.figure(figsize=(18,8))
```

Comparing Examples

```python
#catplot
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", kind="box")
```

```python
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill", order=["Male", "Female"])
```

```python
sns.barplot(
        x="total_bill",
        y="smoker",
        hue="sex",
        data=tips,
        color="pink"
        )
plt.figure(figsize=(16, 8))
```

Compare by Your Code

```python
total=tips.groupby('day')['total_bill'].sum().reset_index() #reset index araştıralım.
```

```python
total
```

```python
smoker=tips[tips.smoker=='Yes']
```

```python
bar1=sns.barplot(x='day',y='total_bill',data=total,color='darkblue')
```

```python
bar2=sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue'
)
```

```python
bar1 = sns.barplot(x="day",  y="total_bill", data=total, color='darkblue')
bar2 = sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue'
)
top_bar = mpatches.Patch(color='darkblue', label='non-smoker')
bottom_bar = mpatches.Patch(color='lightblue', label='smoker = Yes')
plt.legend(handles=[top_bar, bottom_bar])
plt.show()
```

```python
df_iris=sns.load_dataset('iris')
```

```python
df_iris.head()
```

```python
sns.histplot(data=df_iris,x='sepal_length')
```

```python
sns.histplot(data=df_iris,x='sepal_length',kde=True)
```

```python
#REgresyon 
sns.regplot(x=df_iris['sepal_length'], 
            y=df_iris['sepal_width'])
```

```python
#REgresyon with contrast
sns.regplot(
    x=df_iris['sepal_length'], 
    y=df_iris['sepal_width'],
    marker='+',
    line_kws={
                'color':'red',
                'alpha':0.6,
                'lw':5
            },
    scatter_kws={
        'color':'darkblue',
        's':50 #size
    }
            )
```

Compare by Grouped

```python
sns.lmplot(
     x='sepal_length', 
    y='sepal_width',
    data=df_iris,
    hue='species'
)
```

```python
sns.boxplot(x=df_iris['species'],y=df_iris['sepal_length'])
```

```python
sns.boxplot(data=df_iris, x="species", y="sepal_length", hue="species", fill=False, gap=.1)
```

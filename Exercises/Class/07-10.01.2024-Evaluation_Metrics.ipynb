### F1 Score
Precision, Recall https://scikit-learn.org/stable/modules/generated/sklearn.metrics.ConfusionMatrixDisplay.html#sklearn.metrics.ConfusionMatrixDisplay



Predicted--> Kaç tane doğru/ yanlış olacağını tahmin ediyordum.
True labeled--> Predicted için tahminlerimin kaçı doğru kaçı yanlış gibi.
    0,0 --> False pozitive, false tahmin ettim false çıktı aldım.
    0,1->False Negative
    1,1-->True Positive
    1,0-->True Negative


Precision, Recall

Precision is a measure of the accuracy Recall is the true positive rate

Precision = TP / (TP + FP) Recall = TP / (TP + FN)

```python
TP = 10
FN = 2
TN = 12
FP = 1

# class = 1
precision_1 = TP / (TP + FP)
recall_1 = TP / (TP + FN)
```

```python
TP = 12
FN = 1
TN = 10
FP = 2

# class = 0
precision_0 = TP / (TP + FP)
recall_0 = TP / (TP + FN)
```

```python
print(f"class=0 {precision_0:.2f} {recall_0:.2f}")
print(f"class=1 {precision_1:.2f} {recall_1:.2f}")
```

F1-score = 2 * (prc * rec) / (prc + rec)

```python

f1_score_0 = 2 * (precision_0 * recall_0) / (precision_0 + recall_0)
f1_score_1 = 2 * (precision_1 * recall_1) / (precision_1 + recall_1)
```

```python
print(f"{f1_score_0:.2f}")
print(f"{f1_score_1:.2f}")
```

The F1 score's best value is 1.0

- Both Jaccard and F1-score can be used for multi-class classifiers as well.

### Log loss
- If output of classifier is tge probability of a class label instead of the label, we may use log loss. Classifiers with smaller values of log loss are ideal.

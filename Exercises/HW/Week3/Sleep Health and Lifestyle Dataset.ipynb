```python
# Kaggle data set : Sleep Health and Lifestyle Dataset
# https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset/data

""" 
Key Features of the Dataset:

Comprehensive Sleep Metrics: Explore sleep duration, quality, and factors influencing sleep patterns.
Lifestyle Factors: Analyze physical activity levels, stress levels, and BMI categories.
Cardiovascular Health: Examine blood pressure and heart rate measurements.
Sleep Disorder Analysis: Identify the occurrence of sleep disorders such as Insomnia and Sleep Apnea.

Dataset Columns:

Person ID: An identifier for each individual.
Gender: The gender of the person (Male/Female).
Age: The age of the person in years.
Occupation: The occupation or profession of the person.
Sleep Duration (hours): The number of hours the person sleeps per day.
Quality of Sleep (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10.
Physical Activity Level (minutes/day): The number of minutes the person engages in physical activity daily.
Stress Level (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.
BMI Category: The BMI category of the person (e.g., Underweight, Normal, Overweight).
Blood Pressure (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.
Heart Rate (bpm): The resting heart rate of the person in beats per minute.
Daily Steps: The number of steps the person takes per day.
Sleep Disorder: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

Details about Sleep Disorder Column:

None: The individual does not exhibit any specific sleep disorder.
Insomnia: The individual experiences difficulty falling asleep or staying asleep, leading to inadequate or poor-quality sleep.
Sleep Apnea: The individual suffers from pauses in breathing during sleep, resulting in disrupted sleep patterns and potential health risks.
"""
```

Correlation

HeatMap

A heatmap is a graphical representation of data where each value of a matrix is represented as a color with an emphasis on the Seaborn library. Generally speaking, a Pearson correlation coefficient value greater than **0.7** indicates the presence of multi-collinearity.

```python
correlation_matrix = drop_df.corr()

plt.figure()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#Sleep Disorder vs Gender
#Violin Plot
plt.figure(figsize=(15, 6))
plt.xticks(rotation=45, ha='right')
sns.violinplot(x=df["Sleep Disorder"], y=df["Gender"])

"""
 Output : People generally regular type. If we talk about Sleep Apnea and Insomnia  ; Female tend to Sleep Apnea and Insomnia is same .
 I curious about the age , sleep disorder and gender so we can analyze with catplot below.

"""
```

```python
#catplot
sns.catplot(data=df, x="Gender", y="Age", hue="Quality of Sleep", kind="box")
""" 
Output : Older people has better sleep and female aaverage result is high. Additionally, we can say that sleep disorder not affect sleep
quality point directly.

"""
```

```python
GenderS_Gr=df.groupby("Gender")['Sleep Duration'].mean().sort_values(ascending=False)
GenderS_Gr
```

```python
GenderQ_Gr=df.groupby("Gender")['Quality of Sleep'].mean().sort_values(ascending=False)
GenderQ_Gr
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Sleep Duration', data=df, color='pink', label='Sleep Duration')
sns.barplot(x='Gender', y='Quality of Sleep', data=df, color='purple', label='Quality of Sleep')
plt.ylabel('Result')
plt.legend()
plt.show()

# Exactly we can say that sleep duration : female>male
#                         Quality of Sleep : female > male 
```

```python
Occup_S=df.groupby("Occupation")['Sleep Duration'].mean().sort_values(ascending=False)
Occup_S
```

```python

# Matplotlib
plt.figure(figsize=(12, 6))
Occup_S.plot(kind='bar', color='skyblue')
plt.title('Mean Sleep Duration by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Mean Sleep Duration')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
Occup_Q=df.groupby("Occupation")['Quality of Sleep'].mean().sort_values(ascending=False)
Occup_Q
```

```python
#Mean Quality of Sleep by Occupation
# Matplotlib
plt.figure(figsize=(12, 6))
Occup_Q.plot(kind='bar', color='skyblue')
plt.title('Mean Quality of Sleep by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Mean Quality of Sleep')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()

```

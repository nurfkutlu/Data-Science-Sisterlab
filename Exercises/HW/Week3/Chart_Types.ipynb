#### Distribution

##### Histogram

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = sns.load_dataset('iris')
```

```python
df
```

```python
df.info()
```

```python
sns.histplot(data=df, x="sepal_length", kde=True) #Histogram Graph with seaborn
plt.show()
```

##### BoxPlot

```python

sns.boxplot( x=df["species"], y=df["sepal_length"] )
```

##### Density Chart

```python
# Make default density plot
sns.kdeplot(df['sepal_width'], fill=True) #shadow shown at density chart.
```

```python
# plotting both distibutions on the same figure
fig = sns.kdeplot(df['sepal_width'], fill=True, color="r",label='sepal_width')
fig = sns.kdeplot(df['sepal_length'],fill=True, color="b",label='sepal_length')
plt.legend(loc='best') # to see legend
plt.show()
```

##### Violin

```python
# plot
sns.violinplot(x=df["species"], y=df["sepal_length"])
```

#### Ranking

Barplot

```python
sns.barplot(df, x="species", y="sepal_length",label="sepal_length")
sns.barplot(df, x="species", y="sepal_width",label="sepal_width")
plt.legend(loc='best') # to see legend
plt.show()
```

#### Correlation

##### Scatter

```python
# plot
sns.regplot(x=df["sepal_length"], y=df["sepal_width"], line_kws={"color":"r","alpha":0.7,"lw":5}) #line species at line_kws
plt.show()
```

##### Heatmap

```python
 # Each columns should be float / Integer. I drop species from the data frame to solution
drop_df=df.drop("species",axis=1)
correlation_matrix = drop_df.corr()


plt.figure()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title("Correlation Matrix Heatmap")
plt.show()

#
```

##### Bubble Plot

```python
# use the scatterplot function to build the bubble map
sns.scatterplot(data=df, x="petal_width", y="petal_length", size="sepal_width", legend=False, sizes=(20, 2000))

# show the graph
plt.show()
```

##### Connected Scatterplot

```python
sns.lineplot(x="petal_width",
             y="petal_length",
             data=df,
             marker='o', # Style used to mark the join between 2 points
            )
plt.xlabel('petal_width') # x-axis name
plt.ylabel('petal_length') # y-axis name
plt.title('Simple Connected Scatter Plot') # Add a title
plt.show() # Display the graph
```

2D Density Plot

```python
sns.kdeplot(x="sepal_length", y="sepal_width", data=df, fill=True, cmap="Blues", thresh=0.15)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("2D Density Plot for Iris Dataset (Sepal Length vs Sepal Width)")

# Show the plot
plt.show()
```

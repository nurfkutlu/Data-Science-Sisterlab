Define function to obtain dataframe

```python
def book_information(df): 
    for book_info in response['results']['books']:
        Title=book_info['title']
        Publisher=book_info['publisher']
        Description=book_info['description']
        Author=book_info['author']
        Amazon_url=book_info['amazon_product_url']
        First_chapter_Link=book_info['first_chapter_link']
        Age_Group=book_info['age_group']
       
        # save in pandas df
        df = df._append({
            "title": Title,
            "publisher": Publisher,
            "description": Description,
            "author": Author,
            "amazon_product_url":Amazon_url,
            "first_chapter_link":First_chapter_Link,
            "age_group":Age_Group},
        ignore_index = True)
    return df
```

```python
df = pd.DataFrame(columns=["title", "publisher", "description", "author", "amazon_product_url","first_chapter_link","age_group"]) 
Book_info_df = book_information(df)
```
